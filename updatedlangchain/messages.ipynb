# Messages and Prompts in LangChain

## Text Prompt vs. Message Prompt

- **Text Prompt:**
  - A simple string of text given to the language model (e.g., "What is the capital of France?").
  - Used for single-turn, basic interactions.

- **Message Prompt:**
  - A structured format where each message has a role (user, AI, system, etc.) and content.
  - Supports multi-turn conversations and context tracking.

## Types of Messages in LangChain

- **System Message:**
  - Sets the behavior or context for the AI (e.g., "You are a helpful assistant.").

- **Human Message:**
  - Represents input from the user (e.g., a question or instruction).

- **AI Message:**
  - Represents the response generated by the AI model.

- **Function/Tool/Call Message:**
  - Used when the AI needs to call an external function or tool (e.g., a calculator or API call).

## Importance of message
- Message prompts allow for more advanced, context-aware, and interactive conversations.
- Different message types help organize and control the flow of dialogue, making applications more robust and user-friendly.

```python
import os
from langchain.chat_models import init_chat_model
    
# os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')

model = init_chat_model('groq:qwen/qwen3-32b')
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("Your are a Expert of poetry"),
    HumanMessage("Write a poem  on Artifical Inteligence")
]

response = model.invoke(messages)
response.content
```

```python
# Detail Information to the LLM through System Message

from langchain.messages import SystemMessage, HumanMessage, AIMessage
system_message = SystemMessage("""Your are a senior python developer with experties in web framework.
                               always provide response with example.
                               be concise but thorough in you explanation.
                               
                               """)
messages = [
    system_message,  
    HumanMessage("How do i create REST API")
]

response = model.invoke(messages)
print(response.content)
```

```python
human_messagge = HumanMessage(
 content="Hello!",
 name="Faris Ahmed", #Optional 
 id = "msg_123" # Optional
)

response = model.invoke(
    [human_messagge]
)
response.usage_metadata
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

# Create a list of messages
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="What is the capital of France?"),
    AIMessage(content="The capital of France is Paris.")
]

# Pass the messages list to the model
response = model.invoke(messages)
print(response)
```
