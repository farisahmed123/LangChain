# Messages and Prompts in LangChain

## Text Prompt vs. Message Prompt

- **Text Prompt:**
  - A simple string of text given to the language model (e.g., "What is the capital of France?").
  - Used for single-turn, basic interactions.

- **Message Prompt:**
  - A structured format where each message has a role (user, AI, system, etc.) and content.
  - Supports multi-turn conversations and context tracking.

## Types of Messages in LangChain

- **System Message:**
  - Sets the behavior or context for the AI (e.g., "You are a helpful assistant.").

- **Human Message:**
  - Represents input from the user (e.g., a question or instruction).

- **AI Message:**
  - Represents the response generated by the AI model.

- **Function/Tool/Call Message:**
  - Used when the AI needs to call an external function or tool (e.g., a calculator or API call).

## Importance of message
- Message prompts allow for more advanced, context-aware, and interactive conversations.
- Different message types help organize and control the flow of dialogue, making applications more robust and user-friendly.
