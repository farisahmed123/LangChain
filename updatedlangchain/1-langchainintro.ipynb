```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
import os

def get_weather(city: str) -> str:
    """Get the weather for a city."""
    return f"This is the weather of {city} is sunny"
# code
# python
# Load .env and print API key prefix
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")
if api_key:
    print("API key loaded from .env: " + api_key[:6] + '...' if len(api_key) > 6 else api_key)
else:
    print("API key not found in .env")

llm = ChatGroq(
    groq_api_key=os.getenv("GROQ_API_KEY"),
    model_name="llama-70b-versatile"  # updated to supported model
)

agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="You are a good helpful assistant"
)

response = agent.invoke({"input": "What is the weather in New York?"})
print(response)
```
